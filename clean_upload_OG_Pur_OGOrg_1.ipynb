```python
# Import get_dataframe function from the sql module
import pandas as pd 
from sql_functions import get_dataframe
```

```python
schema = 'organic_africa'
df1 = get_dataframe(f'SELECT * FROM {schema}."OG_Pur__OGOrg"') 
```

```python
df1.info()
```

```python
df1['Product'].unique()
```

```python
df1['Product'] = df1['Product'].replace('Paprika fruit whole (org + FT)', 'Paprika fruit whole (org)')
df1.head()
```

```python
df1['Product'] = df1['Product'].replace("Bird's Eye Chilli fruit whole (org + FT)", "Bird's Eye Chilli fruit whole (org)")
df1.head()
```

```python
df1['Product'] = df1['Product'].replace("Rosella (subdariffa) petals (org + FT)", "Rosella (subdariffa) petals (org)")
df1.head()
```

```python
df1['Product'].unique()
```

```python
# we need psycopg2 for raising possible error message
import psycopg2
import sqlalchemy 
from sql_functions import sqlalchemy 
from sql_functions import get_engine 
```

```python
# Write records stored in a dataframe to SQL database
engine = get_engine()
table_name = 'OG_Pur__OGOrg_1'
if engine!=None:
    try:
        df1.to_sql(name='OG_Pur__OGOrg_1', # Name of SQL table variable - at the front its the dataframe!! 
                        con=engine, # Engine or connection
                        schema='organic_africa', # your class schema variable
                        if_exists='replace', # Drop the table before inserting new values 
                        index=False, # Write DataFrame index as a column
                        chunksize=5000, # Specify the number of rows in each batch to be written at a time
                        method='multi') # Pass multiple values in a single INSERT clause
        print(f"The {table_name} table was imported successfully.")
    # Error handling
    except (Exception, psycopg2.DatabaseError) as error:
        print(error)
        engine = None
else:
    print('No engine')
```
