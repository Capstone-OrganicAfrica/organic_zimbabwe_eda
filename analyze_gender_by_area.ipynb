```python
import pandas as pd
import sql_functions as sf
schema = 'organic_africa' # UPDATE 'TABLE_SCHEMA' based on schema used in class 
engine = sf.get_engine()
```

```python
farmers_df = sf.get_dataframe(f'SELECT * FROM {schema}.all_farmers')
farmers_df.shape
```

```python
male_by_area = farmers_df.loc[farmers_df['sex']=='male', 'area'].value_counts()
female_by_area = farmers_df.loc[farmers_df['sex']=='female', 'area'].value_counts()
```

#### Df gender_by_area shows gender by area sorted by asc total

```python
gender_by_area = male_by_area.to_frame()
gender_by_area.columns=['male']
gender_by_area['female'] = female_by_area
gender_by_area.index.names = ['area']
gender_by_area['total'] = female_by_area+male_by_area
gender_by_area['total'] = gender_by_area['total'].astype(int)
gender_by_area['majority'] = gender_by_area[['male','female']].idxmax(axis=1)
gender_by_area.sort_values(by='total', ascending=False, inplace=True)
gender_by_area
```

### Df gender_by_area_percent shows male female distribution by area

Majority of the farmers are **female**! Only in the areas *Mushumbi, Karoi, Triangle* the gender distribution is **50/50**. And *Mbire* is the only area with male majority

```python
def calc_distr(row):
     row["male"] = round(row["male"]/row["total"], 4)
     row["female"] = round(row["female"]/row["total"], 4)
     row["total"] = row["male"]+row["female"]
     return row
```

```python
gender_by_area_percent = gender_by_area.apply(calc_distr, axis=1)
gender_by_area_percent['male'] = gender_by_area_percent['male'].transform(lambda x: '{:,.0%}'.format(x), axis=0)
gender_by_area_percent['female'] = gender_by_area_percent['female'].transform(lambda x: '{:,.0%}'.format(x))
gender_by_area_percent['total'] = gender_by_area_percent['total'].transform(lambda x: '{:,.0%}'.format(x))
gender_by_area_percent.sort_values(by='female', ascending=False, inplace=True)
gender_by_area_percent
```
